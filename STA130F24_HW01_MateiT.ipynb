```python
#Question 5: The difference between the attribute 'df.shape' and the method 'df.describe()' are that df.shape is an attribute and does not need to be called with the parenthesis, this is because there is no computation and a method is not being called on the frame. But rather an attribute of the frame that is already saved. A method like df.describe() must be called with parenthesis due to the possibility of arguments being passed in such as "include = 'objects'", these methods are not handled by the data frame but rather by pandas, and there is commonly computation happening. An easy way to understand this is that attributes store object properties and methods execute operations on the object. 
```
